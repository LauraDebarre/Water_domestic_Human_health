```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
# Set the display option to use scientific notation
pd.set_option('display.float_format', '{:.2e}'.format)
```

```python
# Import CF shapefile
CF=gpd.read_file('../results/CF/CF.shp', driver='ESRI Shapefile') #projection: WGS 84

#change name to clarify : CF reg is just calculated using regional EF, but subriver basin FF
CF=CF.rename(columns={"CF_regiona":"CF_EF_reg"})

#calculate area per subriver basin
CF["cell_area"]=CF.area
CF
```

```python
#import data on water consumption
water_consumption=pd.read_excel('../data/WaterConsumption_2019_notextrapolated_toSend.xlsx', sheet_name="actual_consumption_allsectors")
water_consumption
```

```python
# Calculate the percentage_area of each subiver basin (compared to riverbasin)

# Calculate the sum of cell_area for each BAS34S_ID
CF['sum_cell_area'] = CF.groupby('BAS34S_ID')['cell_area'].transform('sum')

# Calculate the percentage_area
CF['percentage_area'] = CF['cell_area'] / CF['sum_cell_area']
```

```python
#merge CF geodataframe and water use dataframe to quantify water usage per subriver basin
CF_aggregated=pd.merge(CF,water_consumption.rename(columns={"Basin_ID":"BAS34S_ID"})[["BAS34S_ID","Total_cons_m3"]], on="BAS34S_ID", how='left' )

#recalculate water usage per subriver basin (to consider countries) 
CF_aggregated["Conso_subriv_m3"]=CF_aggregated["percentage_area"]*CF_aggregated["Total_cons_m3"]

#Calculate water consumption per region 
CF_aggregated["Conso_country_GLAM_m3"]=CF_aggregated.groupby('ROMNAM')['Conso_subriv_m3'].transform('sum')
CF_aggregated["Conso_Region_WorldBank_m3"]=CF_aggregated.groupby('Region (Wo')['Conso_subriv_m3'].transform('sum')
CF_aggregated["Conso_Region_UN_m3"]=CF_aggregated.groupby('Region (Un')['Conso_subriv_m3'].transform('sum')
CF_aggregated["Conso_Region_WorldBank/UN_m3"]=CF_aggregated.groupby('Region (WB')['Conso_subriv_m3'].transform('sum')
CF_aggregated["Conso_global_m3"]=CF_aggregated['Conso_subriv_m3'].sum()

# Calculate the product of CF and the percentage of conso in the region m3 
CF_aggregated['CF_country_GLAM'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_country_GLAM_m3"]
CF_aggregated['CF_Region_WorldBank'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_Region_WorldBank_m3"]
CF_aggregated['CF_Region_UN'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_Region_UN_m3"]
CF_aggregated['CF_Region_WorldBank/UN'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_Region_WorldBank/UN_m3"]
CF_aggregated['CF_global'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_global_m3"]

#suming the results per region
CF_aggregated['CF_country_GLAM'] = CF_aggregated.groupby("ISO3CD")['CF_country_GLAM'].transform('sum')
CF_aggregated['CF_Region_WorldBank'] = CF_aggregated.groupby("Region (Wo")['CF_Region_WorldBank'].transform('sum')
CF_aggregated['CF_Region_UN'] = CF_aggregated.groupby("Region (Un")['CF_Region_UN'].transform('sum')
CF_aggregated['CF_Region_WorldBank/UN'] = CF_aggregated.groupby("Region (WB")['CF_Region_UN'].transform('sum')
CF_aggregated['CF_global']=CF_aggregated['CF_global'].sum()

CF_aggregated
```

```python
CF_aggregated.to_csv("../results/CF/CF_aggregated_per_regions.csv")
```

```python
#redefining the list of ecoinvent geometries we want 
List_to_complete=ecoinvent_regions[ecoinvent_regions['shortname'].isin(list_needed)]

#redefining the list of ecoinvent COUNTRIES geometries we want according to Maxime's list
List_to_complete_countries=List_to_complete[List_to_complete['collection']=="countries"].rename(columns={"isothreelettercode":"ISO3CD"})

#list of ecoinvent geometries
geographies=['aluminium', 'legacy', 'states', 'electricity',
       'un-subregions', 'special', 'russia', 'un-regions',
       'usa-electricity']

#create a dictionary to complete containing the dataframes with CF aggregated per ecoinvent geographies
dict_df_geographies = {}  # Create an empty dictionary to store DataFrames

for geo in geographies:
    dict_df_geographies['List_to_complete_' + geo] = List_to_complete[List_to_complete['collection'] == geo].rename(columns={"isothreelettercode": "ISO3CD"})
```

```python
#create dictionary with intersection with different geographies and calculate CF per region and geographies

geographies=['aluminium', 'legacy', 'states', 'electricity',
       'un-subregions', 'special', 'russia', 'un-regions',
       'usa-electricity']

intersections_with_geographies={}

for geo in geographies:
    #intersect CF aggregated and 'geo' geographies
    intersections_with_geographies[geo] = gpd.overlay(CF_aggregated, dict_df_geographies['List_to_complete_'+geo], how="intersection")

    #calculate total conso per geo regions
    intersection=intersections_with_geographies[geo]
    intersection["conso_region_"+geo+"_m3"]=intersection.groupby("name")['Conso_subriv_m3'].transform('sum')

    # Calculate the product of CF and the percentage of conso in the region m3 
    intersection["CF_"+geo+"_region"]= intersection['CF']* intersection["Conso_subriv_m3"]/intersection["conso_region_"+geo+"_m3"]

    #suming the results per region
    intersection["CF_"+geo+"_region"]=intersection.groupby("name")["CF_"+geo+"_region"].transform('sum')
    
    #stocking in dict intersection with geographies
    intersections_with_geographies[geo]=intersection

```
