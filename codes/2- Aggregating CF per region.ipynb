```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
# Set the display option to use scientific notation
pd.set_option('display.float_format', '{:.2e}'.format)
pd.set_option('display.max_rows', 30)
```

```python
#Write name of location of the folder 
path_directory="your path to the folder containing the code and data folders"
```

```python
# Import CF shapefile -- The shapefile contains country geometry of GLAM (ISO3CD, ROMNAM)

CF=gpd.read_file(path_directory+'/results/CF/CF.shp', driver='ESRI Shapefile') #projection: WGS 84

#change name to clarify : CF reg is just calculated using regional EF, but subriver basin FF
CF=CF.rename(columns={"CF_regiona":"CF_EF_reg"})

#calculate area per subriver basin
CF["cell_area"]=CF.area
CF
```

```python
#import data on water consumption
water_consumption=pd.read_excel(path_directory+'/data/WaterConsumption_2019_notextrapolated_toSend.xlsx', sheet_name="actual_consumption_allsectors")
water_consumption
```

```python
# Calculate the percentage_area of each subiver basin (compared to riverbasin)

# Calculate the sum of cell_area for each BAS34S_ID
CF['sum_cell_area'] = CF.groupby('BAS34S_ID')['cell_area'].transform('sum')

# Calculate the percentage_area
CF['percentage_area'] = CF['cell_area'] / CF['sum_cell_area']
```

```python
#merge CF geodataframe and water use dataframe to quantify water usage per subriver basin
CF_aggregated=pd.merge(CF,water_consumption.rename(columns={"Basin_ID":"BAS34S_ID"})[["BAS34S_ID","Total_cons_m3"]], on="BAS34S_ID", how='left' )

#recalculate water usage per subriver basin (to consider countries) 
CF_aggregated["Conso_subriv_m3"]=CF_aggregated["percentage_area"]*CF_aggregated["Total_cons_m3"]

#Calculate water consumption per region 
CF_aggregated["Conso_country_GLAM_m3"]=CF_aggregated.groupby('ROMNAM')['Conso_subriv_m3'].transform('sum')   #ROMNAM is the name of the country name in the geography provided by GLAM
CF_aggregated["Conso_Region_WorldBank_m3"]=CF_aggregated.groupby('Region (Wo')['Conso_subriv_m3'].transform('sum')
CF_aggregated["Conso_Region_UN_m3"]=CF_aggregated.groupby('Region (Un')['Conso_subriv_m3'].transform('sum')
CF_aggregated["Conso_Region_WorldBank/UN_m3"]=CF_aggregated.groupby('Region (WB')['Conso_subriv_m3'].transform('sum')
CF_aggregated["Conso_global_m3"]=CF_aggregated['Conso_subriv_m3'].sum()

# Calculate the product of CF and the percentage of conso in the region m3 
CF_aggregated['CF_country_GLAM'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_country_GLAM_m3"]
CF_aggregated['CF_Region_WorldBank'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_Region_WorldBank_m3"]
CF_aggregated['CF_Region_UN'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_Region_UN_m3"]
CF_aggregated['CF_Region_WorldBank/UN'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_Region_WorldBank/UN_m3"]
CF_aggregated['CF_global'] = CF_aggregated['CF'] * CF_aggregated["Conso_subriv_m3"]/CF_aggregated["Conso_global_m3"]

#suming the results per region
CF_aggregated['CF_country_GLAM'] = CF_aggregated.groupby("ISO3CD")['CF_country_GLAM'].transform('sum')
CF_aggregated['CF_Region_WorldBank'] = CF_aggregated.groupby("Region (Wo")['CF_Region_WorldBank'].transform('sum')
CF_aggregated['CF_Region_UN'] = CF_aggregated.groupby("Region (Un")['CF_Region_UN'].transform('sum')
CF_aggregated['CF_Region_WorldBank/UN'] = CF_aggregated.groupby("Region (WB")['CF_Region_UN'].transform('sum')
CF_aggregated['CF_global']=CF_aggregated['CF_global'].sum()

CF_aggregated
```

# ecoinvent regions

```python
#import shapefile for ecoinvent geometries
ecoinvent_regions=gpd.read_file(path_directory+'/data/all.gpkg') #all geographies from the ecoinvent dataset - updated on 06/03/2024
ecoinvent_regions=ecoinvent_regions[["id","gid","collection","name","isothreelettercode","unsubregion","unregion","shortname","geometry"]]
ecoinvent_regions
```

```python
ecoinvent_regions.crs
```

```python
#list of ecoinvent geometries
geographies=['countries','aluminium', 'legacy', 'states', 'electricity',
       'un-subregions', 'special', 'russia', 'un-regions',
       'usa-electricity']

#create a dictionary containing the dataframes with CF aggregated per ecoinvent geographies (10 df)
dict_df_geographies = {}  # Create an empty dictionary to store DataFrames

#for each geography, we proceed to the intersection between the CF (subriver) and the geography
for geo in geographies:
    dict_df_geographies[geo] = ecoinvent_regions[ecoinvent_regions['collection'] == geo].rename(columns={"isothreelettercode": "ISO3CD"})  #we select the geography we want
    dict_df_geographies[geo] = gpd.overlay(CF_aggregated, dict_df_geographies[geo], how="intersection",keep_geom_type=False) #proceed to the intersection 
```

```python
#for each geography, we proceed to the calculation of the % of water consumption of the cell compared to the geographyfor geo in geographies followed by the weighted CF in the cell:

dict_CF=dict_df_geographies

for geo in geographies:
    dict_CF[geo]=dict_CF[geo][['BAS34S_ID', 'Scarcity','XF','EF', 'CF', 'ISO3CD_1', 'ROMNAM','CF_country_GLAM', 'Total_cons_m3','id', 'gid', 'collection', 'name', 'ISO3CD_2','unsubregion', 'unregion', 'shortname','geometry']]
    dict_CF[geo]["cell_area"]=dict_CF[geo].area
   
    # Calculate the sum of cell_area for each BAS34S_ID
    dict_CF[geo]['sum_cell_area'] = dict_CF[geo].groupby('BAS34S_ID')['cell_area'].transform('sum')

    # Calculate the percentage_area
    dict_CF[geo]['percentage_area'] = dict_CF[geo]['cell_area'] / dict_CF[geo]['sum_cell_area']

    #recalculate water usage per subriver basin (to consider countries) 
    dict_CF[geo]["Conso_cell_m3"]=dict_CF[geo]["percentage_area"]*dict_CF[geo]["Total_cons_m3"]
```

```python
#for each geometry, we 1- recalculate the total consumption, 2- Compute CF and percentage of conso in the region

CF_per_geom={}

for geo in geographies:
    dict_CF[geo]["conso_"+geo]=dict_CF[geo].groupby('name')["Conso_cell_m3"].transform('sum')
    dict_CF[geo]["CF_"+geo]=dict_CF[geo]["CF"]*dict_CF[geo]["Conso_cell_m3"]/dict_CF[geo]["conso_"+geo]
    dict_CF[geo]["CF_"+geo]=dict_CF[geo].groupby('name')["CF_"+geo].transform('sum')
 
for geo in ['aluminium', 'legacy', 'states', 'electricity',
       'un-subregions', 'special', 'russia', 'un-regions',
       'usa-electricity']:
    CF_per_geom[geo]=dict_CF[geo][['collection', 'name', 'shortname', 
       'CF_'+geo]]

CF_per_geom["countries"]=dict_CF["countries"][['ISO3CD_1','ROMNAM','CF_country_GLAM','collection', 'name', 'ISO3CD_2',
       'unsubregion', 'unregion', 'shortname', 
       'CF_countries', 'geometry']]
CF_per_geom["countries"]=CF_per_geom["countries"].rename(columns={"ISO3CD_1":"ISO3_GLAM", "ISO3CD_2":"ISO3_ecoinvent"})

```

```python
CF_countries=CF_per_geom["countries"]
```

```python
# Comparison of country geographies : GLAM and ecoinvent
CF_countries=CF_per_geom["countries"]
CF_countries

list_ISO3_GLAM=CF_countries[["ISO3_GLAM", "ROMNAM", "CF_country_GLAM","ISO3_ecoinvent","name","CF_countries"]].drop_duplicates(subset="ISO3_GLAM")
list_ISO3_ecoinvent=CF_countries[["ISO3_GLAM", "ROMNAM", "CF_country_GLAM","ISO3_ecoinvent","name","CF_countries"]].drop_duplicates(subset="ISO3_ecoinvent")

list_ISO3_GLAM_only=CF_countries[["ISO3_GLAM","ROMNAM"]].drop_duplicates(subset="ISO3_GLAM")
list_ISO3_ecoinvent_only=CF_countries[["ISO3_ecoinvent","name"]].drop_duplicates(subset="ISO3_ecoinvent")


list_ISO3_GLAM_only #188 rows
list_ISO3_ecoinvent_only #221 rows


# Convert the columns to sets
set_GLAM = set(list_ISO3_GLAM_only['ISO3_GLAM'])
set_ecoinvent = set(list_ISO3_ecoinvent_only['ISO3_ecoinvent'])

# Find common elements
common_elements = set_GLAM & set_ecoinvent

# Find elements only in df_ISO3_GLAM_only
only_in_GLAM = set_GLAM - set_ecoinvent

# Find elements only in df_ISO3_ecoinvent_only
only_in_ecoinvent = set_ecoinvent - set_GLAM

# Create DataFrames for common and unique elements
df_common = list_ISO3_GLAM_only[list_ISO3_GLAM_only['ISO3_GLAM'].isin(common_elements)]
df_only_in_GLAM = list_ISO3_GLAM_only[list_ISO3_GLAM_only['ISO3_GLAM'].isin(only_in_GLAM)]
df_only_in_ecoinvent = list_ISO3_ecoinvent_only[list_ISO3_ecoinvent_only['ISO3_ecoinvent'].isin(only_in_ecoinvent)]
```
